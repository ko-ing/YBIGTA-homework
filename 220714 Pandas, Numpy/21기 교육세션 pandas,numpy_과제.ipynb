# <과제2>
- 구글링도 과제 수행에 포함되어 있습니다. 세션에 소개되지 않은 기능이 필요하다면, 구글링을 통해 구현해주시길 바랍니다.
- 데이터 : deliver.csv (날씨정보에 따른 배달품목별 주문량 데이터) / '내 드라이브'에 사본 저장할 것.

1. '날짜' 컬럼을 보고, '월' 컬럼과 '계절' 컬럼을 새로 추가하라.
(계절 컬럼 : 3-5월 : 봄, 6-8월 : 여름, 9-11월: 가을, 12-2월 : 겨울)

2. '미세먼지(㎍/㎥)' 컬럼에서 null값의 개수를 확인하고, null값을 전체 평균으로 대체하라.
(전체 평균은 null값이 아닌 값들의 평균으로 계산된다.)

```python
data = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/YBIGTA/교육세션/20220714/deliver.csv")
```

```python
def getMonth(date):
  split = date.split("-")
  return int(split[1])

def getSeason(date):
  month = getMonth(date)
  if (3 <= month and month <=5):
    return "봄"
  elif (6 <= month and month <= 8):
    return "여름"
  elif (9 <= month and month <= 11):
    return "가을"
  else:
    return "겨울"

data['월'] = data['날짜'].apply(getMonth)
data['계절'] = data['날짜'].apply(getSeason)

data.head(30)
```

```python
dirtMean = data.dropna()['미세먼지(㎍/㎥)'].mean()
data['미세먼지(㎍/㎥)'].fillna(dirtMean, inplace=True)

data.head(30)
```
